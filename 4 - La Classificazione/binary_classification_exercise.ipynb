## Soluzione

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
#from sklearn.metrics import plot_roc_curve
```

### Importiamo il dataset

```python
BASE_URL="https://raw.githubusercontent.com/ProfAI/machine-learning-fondamenti/main/datasets/"
df = pd.read_csv(BASE_URL+"breast_cancer.csv")
df.head()
```

```python
df.count()
```

```python
#elimino colonne non utili
df = df.drop("ID number", axis=1)
```

Verifico se i dati sono sbilanciati dal punto di vista delle classi

```python
counts = df["diagnosis"].value_counts()
print(counts)
print(f"Benignant = {counts.iloc[0]} ({counts.iloc[0]/counts.sum()*100:.2f}%) ")
print(f"Malignant = {counts.iloc[1]} ({counts.iloc[1]/counts.sum()*100:.2f}%) ")
```

```python
#effettuo la codifica
map_dict = {"M":1, "B":0}
df["diagnosis"] = df["diagnosis"].map(lambda x: map_dict[x])
df.head()
```

### Creiamo il modello

```python
#features
X = df.drop("diagnosis", axis=1).values
#target
y = df["diagnosis"].values

RANDOM_SEED = 0

#creo training set e test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=RANDOM_SEED)
```

```python
#standardizziamo i dati
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
#i dati sono sbilanciati e quindi il modello sar√† effettuato indicando di bilanciare le classi
#mettendo un peso inversamente proporzionale al numero di osservazioni dell classe
lr = LogisticRegression(class_weight="balanced")
lr.fit(X_train, y_train)
```

### Valutiamo il modello

```python
y_pred_train = lr.predict(X_train)
y_proba_train = lr.predict_proba(X_train)
y_pred_test = lr.predict(X_test)
y_proba_test =  lr.predict_proba(X_test)

print("TRAIN REPORT")
print(classification_report(y_train, y_pred_train))
#accuracy e altre metriche sono intorno al 0,98


print("TEST REPORT")
print(classification_report(y_test, y_pred_test))
#il recall 0.99 non raggiunge le perfomance richieste di 1
#agenfdo sul trashold si corregge la cosa
```

```python
#where permette di impostare un valore o un altro se una certa condizione √® verificata
#in questo caso se y_proba_train √® >0,25 classifica 1 altrimenti 0
y_pred_train = np.where(y_proba_train[:,1]>0.25,1,0)
y_pred_test = np.where(y_proba_test[:,1]>0.25,1,0)
```

```python
print("TRAIN REPORT")
print(classification_report(y_train, y_pred_train))
#accuracy e altre metriche sono intorno al 0,98


print("TEST REPORT")
print(classification_report(y_test, y_pred_test))
```

#Matrice di confusione

```python
def plot_confusion_matrix(y_true, y_pred, labels=["Negative", "Positive"], show_precision_recall=True):

  cm = confusion_matrix(y_true, y_pred) # tn, fp, fn, tp

  df_cm = pd.DataFrame(cm, index = labels,
                    columns = ["Predicted "+labels[0],"Predicted "+labels[1]])
  sns.heatmap(df_cm, annot=True, fmt='g')

  if show_precision_recall:
    # Quale percentuale di classificazione positive erano effettivamente positive ?
    plt.text(0, -0.1, f"Precision: {(cm[1][1]/(cm[1][1]+cm[0][1])):.3f}")
    # Quale percentuale di osservazioni positive sono state classificate come positive ?
    plt.text(1, -0.1, f"Recall: {(cm[1][1]/(cm[1][1]+cm[1][0])):.3f}")


y_pred_train = np.where(y_proba_train[:,1]>0.25,1,0)
plot_confusion_matrix(y_train, y_pred_train, ["Benignant", "Malignant"])
```

```python
y_pred_test = np.where(y_proba_test[:,1]>0.25,1,0)
plot_confusion_matrix(y_test, y_pred_test, ["Benignant", "Malignant"])
```
